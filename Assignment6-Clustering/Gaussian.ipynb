```python
!pip install pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
dp='/content/drive/MyDrive/DM/iris.csv'
iris=pd.read_csv(dp)
```

```python

# Load the iris dataset
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# Creating a Gaussian Mixture Model with 3 components
gmm = GaussianMixture(n_components=3, random_state=0)

# Fitting the model
gmm.fit(data)

# Predicting the cluster labels
labels = gmm.predict(data)

# Plotting the clusters and their centroids
plt.figure(figsize=(8, 6))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=labels, s=40, cmap='viridis')
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], s=100, c='red', label='Centroids')
plt.title('Gaussian Mixture Model Clustering on Iris Dataset')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.legend()
plt.show()
```
