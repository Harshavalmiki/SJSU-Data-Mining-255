```python
#cast timestap as time and not string
df['noted_date']=pd.to_datetime(df['noted_date'])#need to do a manual conversion or else the set up takes too long
```

```python
df.info()
```

```python
#import AnomalyExperiment and initialize its class
from pycaret.anomaly import AnomalyExperiment
exp = AnomalyExperiment()
```

```python
#setup the experiment
exp.setup(df,session_id=123,log_experiment = False)
```

```python
#lets check the list of models we have, that can be used
exp.models()
```

```python
#let us a train a iforest,svm and cluster based outlier
iforest_model = exp.create_model('iforest')
svm_model= exp.create_model('svm')
cluster_model = exp.create_model('cluster')
```

```python
exp.plot_model(iforest_model,plot='tsne')
```

```python
#lets use the iforest_model for making the prediction
iforest_pred = exp.predict_model(iforest_model,data=df)#you can use assign_model as well
```

```python
iforest_pred.head()# we can see that a transform was applied to split the date into year,month and day
```

```python
#savinng the iforest_model
exp.save_model(iforest_model,'/content/drive/MyDrive/iforest')
```
