```python
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/Pycaret_Assignment/current.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#Date column to datetime
data['Datetime']=pd.to_datetime(data['Datetime'])
#Index
data.set_index('Datetime', inplace=True)
```

```python
data.info()
```

```python
#Drop cols
data.drop(['Humidity', 'WindSpeed', 'GeneralDiffuseFlows', 'DiffuseFlows', 'PowerConsumption_Zone2', 'PowerConsumption_Zone3'], axis=1, inplace=True)
```

```python

data.shape
```

```python
#timeseries monthly ferquency
data = data.asfreq('M')
```

```python
data.head()
```

```python
#initialize a time series experiment
e=TSForecastingExperiment()
```

```python
#Setup
e.setup(data=data, target='PowerConsumption_Zone1' ,fh=2, session_id=123,numeric_imputation_target="nearest")
```

```python
#plott
e.plot_model(plot='ts')
```

```python

e.plot_model(plot="diagnostics", fig_kwargs={"height": 800, "width": 1000})
```

```python
#test
e.check_stats(test="summary")
```

```python
#all models
e.models()
```

```python
#create an croston model
croston_model = e.create_model('croston')
```

```python

y_pred = e.predict_model(croston_model)
```

```python
y_pred
```

```python
#plot the sample forecast
e.plot_model(croston_model)
```

```python
#performing tunning on the arima model
tuned_croston_model = e.tune_model(croston_model)
```

```python
#compare
e.plot_model([croston_model, tuned_croston_model], data_kwargs={"labels": ["Baseline", "Tuned"]})
```

```python
#automl to finilize
baseline_models = e.compare_models(n_select=3)
```

```python
#save the metrics
mertics_comparission = e.pull()
```
